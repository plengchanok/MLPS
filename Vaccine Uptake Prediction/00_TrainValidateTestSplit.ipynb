# Train Validation Test Split

```python
import pandas as pd
```

```python
df = pd.read_csv("covidcast_new.csv")

display(df.head(3))
print("Shape: ", df.shape)
```

```python
df['time_value'] = pd.to_datetime(df['time_value'], errors='coerce')
df.info()
```

```python
df = df.sort_values(by = ['time_value','geo_value']).reset_index(drop = True)

display(df.head(3))
print("Shape: ", df.shape)
```

```python
df['time_value'].unique()
```

```python
print("Min Date:", df['time_value'].min())
print("Max Date:", df['time_value'].max())
```

```python
df_train = df[df['time_value']<="2021-01-29"].reset_index(drop = True)
df_valid = df[(df['time_value']>"2021-01-29")&(df['time_value']<="2021-02-05")].reset_index(drop = True)
df_test = df[df['time_value']>"2021-02-05"].reset_index(drop = True)
```

```python
df_train['time_value'].unique()
```

```python
display(df_train.head(3))
display(df_train.tail(3))
print("Shape :", df_train.shape)
```

```python
df_valid['time_value'].unique()
```

```python
display(df_valid.head(3))
display(df_valid.tail(3))
print("Shape :", df_valid.shape)
```

```python
df_test['time_value'].unique()
```

```python
display(df_test.head(3))
display(df_test.tail(3))
print("Shape :", df_test.shape)
```

```python
print("Number of day for train data         :", df[(df['time_value']<="2021-01-29")]['time_value'].nunique())
print("Number of day for validation data    :", df[(df['time_value']>"2021-01-29")&(df['time_value']<="2021-02-05")]['time_value'].nunique())
print("Number of day for test data          :", df[df['time_value']>"2021-02-05"]['time_value'].nunique())
```

```python
print("Data for train data         :", df[(df['time_value']<="2021-01-29")].shape[0])
print("Data for validation data    :", df[(df['time_value']>"2021-01-29")&(df['time_value']<="2021-02-05")].shape[0])
print("Data for test data          :", df[df['time_value']>"2021-02-05"].shape[0])
```

```python
print("%Data for train data         :", round(df[(df['time_value']<="2021-01-29")].shape[0]*100/df.shape[0],4))
print("%Data for validation data    :", round(df[(df['time_value']>"2021-01-29")&(df['time_value']<="2021-02-05")].shape[0]*100/df.shape[0],4))
print("%Data for test data          :", round(df[df['time_value']>"2021-02-05"].shape[0]*100/df.shape[0],4))
```

```python
df_train.to_csv("train_dataset.csv", index = False)
df_valid.to_csv("validation_dataset.csv", index = False)
df_test.to_csv("test_dataset.csv", index = False)
```
